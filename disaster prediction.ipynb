```python
import numpy as np
import pandas as pd
titanic_data=pd.read_csv("titanic-data.csv")
```

```python
titanic_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
titanic_data.hist(bins=100,figsize=(20,15))
plt.show()
```

```python
median=titanic_data["Age"].median()
titanic_data["Age"].fillna(median)
```

```python
titanic_data.head(1)
```

```python
titanic_data.drop(["PassengerId","Name","Ticket","Cabin"],axis=1,inplace=True)
titanic_data.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
titanic_data["Sex"]=labelencoder.fit_transform(titanic_data["Sex"])
```

```python
titanic_data1=pd.get_dummies(titanic_data)
median=titanic_data1["Age"].median()
titanic_data1["Age"]=titanic_data1["Age"].fillna(median)
titanic_data1.head(20)
```

```python
X=titanic_data1.drop("Survived",axis=1)
X
```

```python
Y=titanic_data1["Survived"]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion="entropy",max_depth=4)
model.fit(X_train,Y_train)
model
```

```python
model_pr=model.predict(X_test)
from sklearn.metrics import f1_score
print("Decision tree f1-score: %0.2f" % f1_score(model_pr,Y_test))
```

```python
plt.scatter(x=model_pr,y=Y_test,alpha=0.1)
plt.show()
```
